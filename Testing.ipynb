```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')

x = np.arange(-2,2,.1)
y = np.arange(-2, 2, .1)

X,Y = np.meshgrid(x,y)

Z = X**2 + Y**2 + 2*X - 3*Y

surf = ax.plot_surface(X,Y,Z)

plt.show()
```
